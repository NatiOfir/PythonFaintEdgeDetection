```python
import numpy as np
!pip install open
import cv2
import time
from queue import PriorityQueue
from pathlib import Path
import math

PI = 3.141592653589793238463

def println(val):
    print(val)

def tic():
    return time.time()

def toc(start):
    elapsed = (time.time() - start)
    println(f"Elapsed {elapsed:.10f} Seconds")
    return elapsed

def end_run(exit_code):
    input()
    return exit_code

def sign(x):
    if x > 0:
        return 1
    elif x < 0:
        return -1
    return 0

def nchoosek(n, k):
    if k == 0:
        return 1
    if n == 0:
        return 0
    return nchoosek(n - 1, k - 1) + nchoosek(n - 1, k)

def sub2ind(rows, cols, row, col):
    return cols * row + col

def mat_sub2ind(rows, cols, row, col):
    """Convert matrix subscripts to linear indices"""
    assert row.shape == col.shape
    return cols * row + col

def ind2sub(ind, cols, rows):
    """Convert linear index to matrix subscripts"""
    row = ind // cols
    col = ind % cols
    return row, col

def exists(filepath):
    """Check if file exists"""
    return Path(filepath).exists()

def read_image(img_path, kill=False):
    """Read image in grayscale"""
    image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        println("Could not open or find the image")
        if kill:
            exit(-1)
    return image

def show_image(image, fig, scale=1, wait=False):
    """Display image with optional scaling"""
    i_big = cv2.resize(image, (0, 0), fx=scale, fy=scale)
    win_name = f"Window {fig}"
    cv2.namedWindow(win_name, cv2.WINDOW_AUTOSIZE)
    cv2.imshow(win_name, i_big)
    if wait:
        cv2.waitKey(0)

def find_indices(M):
    """Find non-zero indices in matrix"""
    return np.nonzero(M.ravel())[0]

def copy_indices(D, ind):
    """Copy values from D at indices specified by ind"""
    values = D.ravel()
    dest = np.zeros_like(ind, dtype=float)
    mask = ind >= 0
    dest[mask] = values[ind[mask]]
    return dest

def set_value_if_true(value, dst, flag):
    """Set values in dst where flag is True"""
    if isinstance(value, (int, float)):
        dst[flag] = value
    else:
        dst[flag] = value[flag]

def keep_selected_rows(src, good_cols):
    """Keep only rows where good_cols is True"""
    return src[good_cols.ravel(), :]

def keep_selected_columns(src, good_cols):
    """Keep only columns where good_cols is True"""
    return src[:, good_cols.ravel()]

def keep_true(src, keep):
    """Keep only elements where keep is True"""
    return src[keep]

def set_values_in_ind(values, ind, dst):
    """Set values at specified indices in dst"""
    dst.ravel()[ind] = values

def max_value(m):
    """Get maximum value in matrix"""
    return np.max(m)

def get_highest_k_values(src, k):
    """Get k highest values and their indices"""
    flat_src = src.ravel()
    idx = np.argpartition(flat_src, -k)[-k:]
    idx = idx[np.argsort(-flat_src[idx])]
    return flat_src[idx], idx

def reorder(idx, src):
    """Reorder elements according to indices"""
    return src.ravel()[idx].reshape(src.shape)

def reorder_cols(idx, src):
    """Reorder columns according to indices"""
    return src[:, idx]

def ind_to_angle(rows, cols, ind0, ind1):
    """Convert indices to angle"""
    x0, y0 = ind2sub(ind0, cols, rows)
    x1, y1 = ind2sub(ind1, cols, rows)
    v1 = x1 - x0
    v2 = y1 - y0
    
    if v1 == 0:
        angle = sign(v2) * PI / 2
    else:
        angle = math.atan(v2 / v1)
    
    angle *= 180 / PI
    ang = int(angle) + 360
    
    if v1 < 0:
        ang += 180
    ang = ang % 360
    
    assert 0 <= ang < 360
    return ang

def ind_to_angle_mat(rows, cols, ind0, ind1):
    """Convert matrices of indices to angles"""
    angles = np.zeros_like(ind0)
    for i in range(ind0.size):
        angles.ravel()[i] = ind_to_angle(rows, cols, int(ind0.ravel()[i]), int(ind1.ravel()[i]))
    return angles
```

```python
def run_im(self) -> np.ndarray:
        """Edge Detection Runner"""
        m = self._handle.m
        n = self._handle.n
        
        if self._prm.printToScreen:
            print("Build Binary Tree")
            start = time.time()
            
        S = np.arange(m*n, dtype=TYPE).reshape(m,n)
        self._handle.S = S.copy()
        self.beam_curves(0, 0, S)
        
        if self._prm.printToScreen:
            print(f"Time: {time.time() - start:.2f}s")
            print("Create Edge Image")
            start = time.time()
            
        self.get_scores()
        
        if self._prm.printToScreen:
            print(f"Time: {time.time() - start:.2f}s")
            
        return self._E

    def beam_curves(self, index: int, level: int, S: np.ndarray):
        """Construct the Beam-Curve Binary-Tree"""
        m, n = S.shape

        if max(m,n) <= self._prm.patchSize:
            self._maxLevel = level
            self.get_bottom_level_simple(S, index)
        else:
            vertical_split = n >= m
            
            if vertical_split:
                mid1 = n // 2
                mid2 = m // 2
                S0 = S[:, :mid1].copy()
                S1 = S[:, mid1:].copy()
                S2 = S[:mid2, :].copy()
                S3 = S[mid2:, :].copy()
            else:
                mid1 = m // 2
                mid2 = n // 2
                S0 = S[:mid1, :].copy()
                S1 = S[mid1:, :].copy()
                S2 = S[:, :mid2].copy()
                S3 = S[:, mid2:].copy()

            t = [2 * index + 1, 2 * index + 2]
            S_arr1 = [S0, S1]

            if self._prm.parallel and level % self._prm.parallelJump == 0:
                threads = []
                for i in range(2):
                    thread = Thread(target=self.beam_curves, 
                                 args=(t[i], level + 1, S_arr1[i]))
                    threads.append(thread)
                    thread.start()
                for thread in threads:
                    thread.join()
            else:
                for i in range(2):
                    self.beam_curves(t[i], level + 1, S_arr1[i])

            try:
                np.maximum(self._pixelScores[t[0]], self._pixelScores[t[1]], 
                          out=self._pixelScores[index])
            except:
                self._pixelScores[index] = np.maximum(self._pixelScores[t[0]], 
                                                    self._pixelScores[t[1]])

            if self._prm.minLevelOfStitching == 0 or level >= self._prm.minLevelOfStitching:
                self.merge_tiles_simple(S0, S1, index, level, vertical_split)
                if vertical_split:
                    self.merge_tiles_simple(S2, S3, index, level, not vertical_split)

            # Merge data from children
            for i in range(2):
                for key, value in self._data[t[i]].items():
                    self.insert_value_to_map(index, float(key), value)
                self._data[t[i]].clear()

    def insert_value_to_map(self, index: int, key: float, value: np.ndarray) -> bool:
        """Insert a value into the data map at the specified index"""
        if not self._data[index]:
            self._data[index] = {}
            
        if key not in self._data[index]:
            self._data[index][key] = value.copy()
            return True
            
        if value[self._handle.SC] > self._data[index][key][self._handle.SC]:
            self._data[index][key] = value.copy()
            return True
            
        return False

    def merge_tiles_simple(self, S1: np.ndarray, S2: np.ndarray, index: int, level: int, vertical_split: bool):
        """Merge two sibling tile responses"""
        edge_S1 = self.get_edge_indices(S1)
        edge_S2 = self.get_edge_indices(S2)
        split = np.array([], dtype=TYPE)

        if vertical_split:
            self.get_best_splitting_points(edge_S1[2], split, index)
            
            for i in range(len(edge_S1)):
                if i == 2:  # skip split of left rectangle
                    continue
                self.find_best_response(edge_S1[i], split, edge_S1[i], index, level)
                
                for j in range(len(edge_S2)):
                    if j == 0 or i == j:  # skip split of right rectangle and same sides
                        continue
                    self.find_best_response(edge_S1[i], split, edge_S2[j], index, level)
        else:
            self.get_best_splitting_points(edge_S1[3], split, index)
            
            for i in range(len(edge_S1)):
                if i == 3:  # skip split of upper rectangle
                    continue
                self.find_best_response(edge_S1[i], split, edge_S1[i], index, level)
                
                for j in range(len(edge_S2)):
                    if j == 1 or i == j:  # skip split of lower rectangle and same sides
                        continue
                    self.find_best_response(edge_S1[i], split, edge_S2[j], index, level)

    def get_edge_indices(self, S: np.ndarray) -> List[np.ndarray]:
        """Returns the indices of the tile edges"""
        edges = []
        edges.append(S[:, 0].reshape(-1))    # left edge
        edges.append(S[0, :].reshape(-1))    # top edge
        edges.append(S[:, -1].reshape(-1))   # right edge
        edges.append(S[-1, :].reshape(-1))   # bottom edge
        return edges
```

```python
import numpy as np
import cv2
import time
from queue import PriorityQueue
from threading import Lock, Thread
import math
from dataclasses import dataclass
from typing import List, Dict, Tuple
import matplotlib.pyplot as plt

PI = math.pi
CIRCLE = 360
TYPE = np.float64
ZERO = 0.0
FALSE = False
BOOL = bool

@dataclass
class Handle:
    m: int = 0  # rows
    n: int = 0  # cols 
    N: int = 0  # total pixels
    S: np.ndarray = None
    rSize: Tuple[int, int] = (0, 0)
    TOTAL: int = 10  # Total number of fields in data structure
    # Data structure field indices
    C: int = 0    # Contrast
    I0S0: int = 1 # First subsegment index
    L: int = 2    # Length
    R: int = 3    # Response
    S0I1: int = 4 # Second subsegment index
    SC: int = 5   # Score
    minC: int = 6 # Min contrast
    maxC: int = 7 # Max contrast
    A0: int = 8   # Start angle
    A1: int = 9   # End angle

@dataclass
class MyParam:
    filterWidth: int = 2
    fibers: bool = False
    patchSize: int = 16
    parallel: bool = True
    parallelJump: int = 2
    minLevelOfStitching: int = 0
    printToScreen: bool = True
    normType: int = 2
    interpolation: bool = True
    removeEpsilon: float = 1.0
    sigma: float = 1.0
    minContrast: float = 2.0
    maxTurn: int = 45
    splitPoints: int = 3
    nmsFact: float = 0.6
    maxNumOfEdges: int = 1000

def read_image(img_path: str, kill: bool = False) -> np.ndarray:
    """Read image in grayscale"""
    image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print("Could not open or find the image")
        if kill:
            exit(-1)
    return image

def show_image(image: np.ndarray, fig: int, scale: float = 1, wait: bool = False):
    """Display image with matplotlib"""
    plt.figure(fig)
    plt.imshow(image, cmap='gray')
    plt.axis('off')
    plt.show()
    if wait:
        plt.pause(0.1)

class Detector:
    def __init__(self, I: np.ndarray, prm: MyParam):
        self._prm = prm
        self._w = prm.filterWidth
        self._debug = False
        self._maxLevel = 0
        self._mtx = Lock()
        self._handle = Handle()
        self._pixels = {}  # Dict[int, np.ndarray]
        
        # Create filter
        if not prm.fibers:
            filter = np.zeros(2 * self._w + 1, dtype=TYPE)
            filter[:self._w] = 1
            filter[self._w+1:] = -1
        else:
            filter = np.zeros(2 * self._w + 1, dtype=TYPE)
            filter[:self._w] = -1
            filter[self._w+1:] = -1
            filter[self._w] = self._w * 2
            filter /= 2
            
        # Convert image and filter
        self._I = I.astype(TYPE)
        self._filter = filter.astype(TYPE)
        self._E = np.zeros_like(self._I, dtype=TYPE)
        
        # Calculate gradients
        self._dX = cv2.filter2D(self._I, -1, self._filter, borderType=cv2.BORDER_REFLECT)
        self._dY = cv2.filter2D(self._I, -1, self._filter.reshape(-1,1), borderType=cv2.BORDER_REFLECT)
        
        # Initialize handles
        m, n = self._I.shape
        N = n * m
        self._handle.m = m
        self._handle.n = n
        self._handle.N = N
        self._handle.rSize = (N, N)
        
        # Initialize data structures
        max_size = int(np.ceil(N/3.0))
        self._pixelScores = [np.zeros((m,n), dtype=TYPE) for _ in range(max_size)]
        self._data = [{} for _ in range(max_size)]
        
        if self._debug:
            print(self._I)
            print(self._filter)
            print(self._dY)
            print(self._dX)
```
